```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
main idea of simulation : 大數法則(經驗分配)逼近母體分配

ex: 高維微積分 (複雜) 運算速度慢

數學方法:黎曼和
統計方法: 聯合機率分布。用估計的

'''
```

```python
u = []
u2 = []
v = []

for i in range(100000):
    num = random.uniform(0,1)
    u.append(num)
    u2.append(num**2)
    v.append((1 - num**2)**0.5)
```

```python
corr_matrix = np.corrcoef(u, v)
corr_matrix
```

```python
corr_matrix = np.corrcoef(u2, v)
corr_matrix
```

```python
import matplotlib.pyplot as plt

plt.scatter(u2, v, alpha=0.05, s=1)
plt.xlabel("U^2")
plt.ylabel("sqrt(1 - U^2)")
plt.title("Scatter plot of (U^2, sqrt(1 - U^2))")
plt.grid(True)
plt.show()
```

```python
'''
這段模擬實際是建立在圓內的點滿足 
𝑥=𝑈,𝑦=1−𝑈2

可看作是將 U∈[0,1] 映射到單位圓的右半邊。
'''
```

```python
def getN(n):
    x = []
    for i in range(n):
        num = []
        while True:
            num .append(random.uniform(0,1))
            if sum(num) > 1:
                x.append(len(num))
                break
                
    return x
```

```python
t = getN(1000000)
sum(t) / len(t)
```

```python

'''
 這個過程的隨機變數 N 的機率期望是：
𝐸[𝑁]=𝑒≈2.71828
'''
```
