```python
'''

前提: 抽樣誤差: 可容忍上界和n有關


柴比雪夫 : P (x_bar - theta <= k*σ / n**0.5) >= 1-(1/k**2)

n >= σ**2 / a**2 * e**2 。 σ小 ， n可以抽比較少


中央極限定理 (n ==> inf)




problem1 : 實務上n永遠不可能接近inf

problem2 : σ未知(通常用s估計)



當分布無法趨近normal or  σ求不出來

直接測E[x] by重複試驗，看是否平穩 (simulation)。

problem 只有一組 X1.....xn，如何重複試驗? ==> bootstrapping(resampling technique)

main idea: 用一組資料模擬新data，來重複試驗。 

bunch of (x1....xn) ~ F_hat!

'''
```

```python
'''
x1, x2,.................xn

Jackknife : drop 1 data for each。當n夠大，還是相同分配。(without replacement)(無母數)

(x2,x3..xn)  (x1,x3...xn) .......(x1....xn-1)

問題: F_hat(平行世界)數量固定(n組)，沒有隨機性。

當theta_hat有outlier，remove!



'''
```

```python
'''
x1, x2,.................xn

bootstrapping : 隨機重抽B次。(with replacement)(無母數)

(x1...xn)1  (x1...xn)2 .......(x1...xn)B

n, B ==> inf。


'''
```

```python
'''
x1, x2,.................xn

paramatric bootstrapping : F已知。

(x1...xn)1  (x1...xn)2 .......(x1...xn)B

直接把theta_hat塞回F。F(x|theta_hat) == F_hat(x)


'''
```

step1: compute theta_hat(sample mean) based on 題目給定samples(x1......x10)

step2: (Resampling) generate (x'1 ....x'10) from F(x | sample mean) 。將sample mean直接帶入常態，再從常態抽取新樣本

step3: generate 1000(b)個平行宇宙，每個宇宙裡面有10(n)個new samples (x'1.....x'10)

step4: compute 1000 trimmed mean from 1000個平行宇宙

step5: compute standard error

```python
import numpy as np
import random
```

```python
samples=[78,86,97,91,83,89,92,88,79,68]  #題目給定的(x1......x10)
```

```python
sample_mean=np.mean(samples) #step1 計算sample mean
print(sample_mean)
```

```python
samples.sort()
sample_trimmed_mean=sum(samples[2:8])/6
print(sample_trimmed_mean)
```

```python
# F已知為 N ~(u,8)。
```

```python
random.seed(0)
#resampling
n=10
b=1000
List=[]
for i in range(b): #step3 生成F，每個F裡面10個新樣本
    List.append([])
    for j in range(n):
        List[i].append(np.random.normal(sample_mean ,8))  #step2 從 N(samples_mean ,8) 生成新樣本
```

```python
#step4 計算每個宇宙的trimmed mean

   
a=[]
for i in range(b):
    List[i].sort
    a.append(sum(List[i][2:8])/6)
```

```python
len(a)
```

```python
#step5 計算stde
a=np.array(a)
stde=np.std(a,ddof=1)
print(stde)
```

```python
'''
bootstrapping大誤區: theta_hat計算stde，不須除以n**0.5 (每個F直接估計母數了)。若n ==>inf，theta_hat固定(無法計算)。
'''
```

```python
#無母數bootstrapping
```

```python
#resampling
random.seed(0)
List=[]
for i in range(b): #step2 生成1000個平行宇宙，每個宇宙裡面10個 新樣本
    List.append([])
    for j in range(n):
        List[i].append(np.random.choice(samples)) #step 1 從題目給定的樣本中，取後放回抽出新樣本
       
```

```python
#step3 計算每個宇宙的trimmed mean
a=[]
for i in range(b):
    List[i].sort()
    a.append(sum(List[i][2:8]))
    a[i]=a[i]/6


 
```
