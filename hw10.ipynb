```python
import random
import numpy as np
import math
```

題目：檢測data=[122,133,106,128,135,126]是否為從指數分配中取出？？

h0:該組樣本符合指數分配 h1:該組樣本不符合指數分配

題目沒有給出任何母數(Goodness of Fit Tests When Some Parameters Are Unspecified)

採用bootstarp 加上ks檢定，計算Ｄ (KS檢定統計量)，並求出p value

 若Ｄ很小，表示經驗分配跟目標很相近。符合h0

D = Maximum{j/n − F(y( j)), F(y( j)) − j-1/n }

Algorithm

Step1:從題目給的dataset，估計MLE(theta hat)，並算出Ｄ

step2：bootstrapping(有母數)。生成10000個平行宇宙，每個平行宇宙裡面6個樣本。樣本取自指數分配 (lambda=1/theta hat(主宇宙求出))

step3:每個平行宇宙內，重新估計各自MLE，並算出各自的d。

step4:將10000個d進行排序，將主宇宙的Ｄ插入，求出p value

```python
data=[122,133,106,128,135,126]

theta=sum(data)/len(data)

print('theta hat =',theta)
```

```python
# CDF of exp(theta_hat)

y = sorted(data)

Fy = []

for i in range(6):
    Fy.append(1-(math.exp(-y[i]/theta)))
```

```python
a = []
b = []
for i in range(1,7):
    a.append(i/6)
    b.append((i-1)/6)
a=np.array(a) #題目n=6 j=1,2...6 ，j/n求出 a= [1/6,2/6 ........1]

b = np.array(b)
```

```python
L = []
for i in range(len(Fy)):
    num_1=a[i]-Fy[i]
    num_2=Fy[i]-b[i]
    L.append(max(num_1, num_2))
D=max(L)
print('D = ',D) 
```

```python
#step2：bootstrapping(有母數)。

random.seed(0)
k=10000
b=6
sim=[]
for i in range(k):#生成10000個 F
    sim.append([])
    for j in range(b):#每個F裡面6個樣本，有母數bootstrapping (exp)
        sim[i].append(-theta*math.log(random.uniform(0,1))) 
```

```python
#將每個F的樣本進行排序
for i in range(k):
    sim[i]=sorted(sim[i])
```

```python
# step3 每個F計算各自的 MLE，並將samples丟入各自的 F() ，之後算出各自的d


def func(param):
    theta=np.mean(param)
    retval = (1-np.exp(-param/theta))
    return retval

sim = np.array(sim)
y = []
for i in range(k):
    y.append([])
    y[i] = func(sim[i])


```

```python
y=np.array(y) #轉為np，方便運算
```

```python
#computing 各自F的d。
List = []
D = []

for i in range(k):
    num1 = a - y[i]
    num2 = y[i] - b
    d=max(np.append(num1,num2))
    D.append(d)
```

```python
#step 4 插入原本的Ｄ 並將其排序，查看位子。計算p value
D.append(0.5717293827873403) #
D=sorted(D)
```

```python
import pandas as pd
df=pd.DataFrame(D)
df['percentile']=df[0].rank(pct=True)
```

```python
#計算p(d>D) d是一整個分配 前面算的D是檢定統計量。如果d>D就=1 否則=0 
#p value = 總和/模擬次數
#p value=0.0007 <0.05 。因此reject h0 ， 題目給的這組dataset不服從指數分配
```

```python
list=[]
for i in range(k):
    if D[i]>0.5717293827873403 :
        list.append(1)
    else:
        list.append(0)
p_value = sum(list)/k    
print('p_value;',p_value)
```
